area of triangle

```
# area of triangle
a = 2
b = 3
c = 5

s = a+b+c/2

area = (s*(s-a)*(s-b)*(s-c))**0.5
print("area of triangle is ", area)
```

Quadratic equation

```
# quadratic equation
import cmath
a = 2
b = 3
c = 5

d = (b ** 2) - (4*a*c)

q1 = (-b + cmath.sqrt(d))/(2*a)
q2 = (-b - cmath.sqrt(d))/(2*a)
print("solution of 1st and second: ", q1,"",q2)
```

### Basic `for` loop example

```
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  print(x)
```

### Basic `while` loop example

```
count = 0
while count < 5:
  print(f"Count is: {count}")
  count += 1
print("Loop finished!")
```

### Basic `break` statement example

```
for i in range(10):
  if i == 5:
    print(f"Breaking the loop at i = {i}")
    break
```

### Basic continue statement


```
for i in range(10):
  if i == 5:
    print(f"Breaking the loop at i = {i}")
    continue
```

### Set Data Type in Python

Sets are unordered collections of unique elements. This means that each element within a set must be unique, and sets do not maintain any specific order. They are defined by enclosing elements in curly braces `{}` or by using the `set()` constructor.

```
s = {1, 2, 3, 4, 5, 3, 2}
print("My set:", s)
print("Type of my_set:", type(s))
```
